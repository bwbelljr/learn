Imagine we are interested in predicting penguins based on two of their body measurements: culmen length and culmen depth. First we want to do some data exploration to get a feel for the data.

What are the features? What is the target?

Features - culmen length, culmen depth
Target - penguin species

The data is located in `../datasets/penguins_classification.csv`, load it with `pandas` into a `DataFrame`.

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/master/datasets/penguins_classification.csv')
```
