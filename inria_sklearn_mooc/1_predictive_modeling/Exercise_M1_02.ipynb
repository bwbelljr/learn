Load adult census dataset with only numeric variables.

```python
import pandas as pd
adult_census = pd.read_csv("https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/master/datasets/adult-census-numeric.csv")
data = adult_census.drop(columns=["class"])
target = adult_census["class"]
```

One of the `KNeighborsClassifier` parameters is `n_neighbors`, which controls the number of neighbors we are going to use to make a prediction for a new data point.

What is the default value of the `n_neighbors` parameter? 5

```python
from sklearn.neighbors import KNeighborsClassifier
```

Create a `KNeighborsClassifier` model with `n_neighbors=50`

```python
model = KNeighborsClassifier(n_neighbors=50)
```

Fit this model on the data and target loaded above

```python
model.fit(data, target)
```

Use your model to make predictions on the first 10 data points inside the data. Do they match the actual target values?

```python
first_ten_preds = model.predict(data[:10])
first_ten_preds
```

```python
target[:10]
```

```python
(first_ten_preds == target[:10]).sum()
```

Predictions do not perfectly match targets.

Compute accuracy on training data

```python
model.score(data, target)
```

Now load the test data and compute the accuracy on the test data

```python
adult_test = pd.read_csv("https://raw.githubusercontent.com/INRIA/scikit-learn-mooc/master/datasets/adult-census-numeric-test.csv")
data_test = adult_test.drop(columns=['class'])
target_test = adult_test['class']
```

```python
model.score(data_test, target_test)
```

Accuracy seems slightly higher with `n_neighbors=50` than with `n_neighbors=5` (default value)
